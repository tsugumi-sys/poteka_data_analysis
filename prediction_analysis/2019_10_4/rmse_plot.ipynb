```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

ten_minutes_mlruns_folder = "/home/akira/Desktop/p-poteka/poteka_pipeline/mlruns/1/9dcd580a6c1d4fbbb69cc4b3a49814d9/artifacts/evaluations/Sequential_NOT_TC_case_2019-10-04_5-0_start/"
sixty_minutes_mlruns_folder = "/home/akira/Desktop/p-poteka/poteka_pipeline/mlruns/1/9dcd580a6c1d4fbbb69cc4b3a49814d9/artifacts/evaluations/NOT_TC_case_2019-10-04_5-0_start/"

assert os.path.exists(ten_minutes_mlruns_folder)
assert os.path.exists(sixty_minutes_mlruns_folder)
```

```python
ten_rmse = [2.052, 3.181, 4.524, 3.881, 4.021, 3.701]
one_hor_rmse = [1.403, 1.110, 1.451, 1.083, 1.204, 1.271]
time_steps = ["5:00UTC", "5:10UTC", "5:20UTC", "5:30UTC", "5:40UTC", "5:50UTC"]

df = pd.DataFrame(columns=["ten_m", "one_h"], index=time_steps)

for idx, rmse in enumerate(ten_rmse):
    df.loc[time_steps[idx], "ten_m"] = rmse

for idx, rmse in enumerate(one_hor_rmse):
    df.loc[time_steps[idx], "one_h"] = rmse

df.head()
```

```python
ax = sns.lineplot(data=df)
ax.set_title("RMSE of 1 hour and 10 minutes prediction of 2019/10/04 5:00UTC ~ 5:50UTC.")
ax.set_ylabel("RMSE (mm/h)")
ax.set_xlabel("Time")
plt.tight_layout()
plt.savefig("2019_10_04.png")
plt.show()
plt.close()
```
