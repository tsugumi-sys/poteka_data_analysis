```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

ten_minutes_mlruns_folder = "/home/akira/Desktop/p-poteka/poteka_pipeline/mlruns/1/9dcd580a6c1d4fbbb69cc4b3a49814d9/artifacts/evaluations/Sequential_TC_case_2020-10-12_6-0_start/"
sixty_minutes_mlruns_folder = "/home/akira/Desktop/p-poteka/poteka_pipeline/mlruns/1/9dcd580a6c1d4fbbb69cc4b3a49814d9/artifacts/evaluations/TC_case_2020-10-12_6-0_start/"

assert os.path.exists(ten_minutes_mlruns_folder)
assert os.path.exists(sixty_minutes_mlruns_folder)
```

```python
ten_rmse = [2.145, 3.202, 2.785, 4.143, 4.503, 5.309]
one_hor_rmse = [2.145, 7.526, 11.896, 18.304, 25.282, 30.310]
time_steps = ["7:00UTC", "7:10UTC", "7:20UTC", "7:30UTC", "7:40UTC", "7:50UTC"]

df = pd.DataFrame(columns=["ten_m", "one_h"], index=time_steps)

for idx, rmse in enumerate(ten_rmse):
    df.loc[time_steps[idx], "ten_m"] = rmse

for idx, rmse in enumerate(one_hor_rmse):
    df.loc[time_steps[idx], "one_h"] = rmse

df.head()
```

```python
ax = sns.lineplot(data=df)
ax.set_title("RMSE of 1 hour and 10 minutes prediction of 2020/10/12 7:00UTC ~ 7:50UTC.")
ax.set_ylabel("RMSE (mm/h)")
ax.set_xlabel("Time")
plt.tight_layout()
plt.savefig("2020_10_12.png")
plt.show()
plt.close()
```
