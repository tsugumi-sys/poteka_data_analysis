```python
import os
import pandas as pd
import numpy as np
from datetime import datetime, timedelta, date
import matplotlib.pyplot as plt
import seaborn as sns
import re
import sklearn

# [RAF].rolling(60).sum()
root_path = '../../../data/imputed-data/Vlz-Bagbaguin_00174731/2020/07'
#'../../../data/cleaned-data/Vlz-Bagbaguin_00174731/2020/07/2020-07-21/data.csv'
today = '2020-07-21'
yesterday = date.fromisoformat(today) - timedelta(days=1)
df_today = pd.read_csv(root_path + f'/{today}/data.csv', index_col='Datetime')
yesterday_path = root_path + f'/{yesterday}/data.csv'
if os.path.exists(yesterday_path):
    df_yesterday = pd.read_csv(yesterday_path, index_col='Datetime')
    df = pd.concat([df_yesterday, df_today])
    df["hour-rain"] = df["RAF"].rolling(60, min_periods=1).sum()
    df_today["hour-rain"] = df["hour-rain"].loc[df_today.index]
    df_today.to_csv('test.csv')
    
df.tail()
```
