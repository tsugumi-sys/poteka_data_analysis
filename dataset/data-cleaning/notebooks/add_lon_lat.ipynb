```python
import os
import pandas as pd
import numpy as np
import re

dates = ['2020-04-18']
#folder_path = '../../../data/accumulated-raf-data'
def find_csv_file(path):
    files = os.listdir(path)
    result = []
    for file in files:
        for item in re.findall('^hk.+.csv$', str(file)):
            result.append(item)
    if len(result) > 1:
        return result[-1]
    else:
        return result[0]
    return result


def find_value(x):
    result = re.findall('\d+', str(x))
    if len(result) > 1:
        return result[-1]
    elif len(result) == 1:
        return result[0]
    else:
        return np.nan

folder_path = '../../../data/poteka-raw-data/'
cols = ["LON", "LAT"]
ob_position = pd.DataFrame(columns=["LON", "LAT"])
ob_folders = os.listdir(folder_path)
for ob_folder in ob_folders:
    folders = os.listdir(folder_path + f'/{ob_folder}')
    year = '2020'
    month = '04'
    date = dates[0]
    path = folder_path + f'/{ob_folder}/{year}/{month}/{date}/HK/'
    if os.path.exists(path):
        file_name = find_csv_file(path)
        if os.path.exists(path + file_name):
            print(path, file_name)
            df = pd.read_csv(path + file_name)
            for col in cols:
                for col_def in df.columns:
                    if col in str(df[col_def].values[0]):
                        df[col] = df[col_def].apply(find_value)
                        df[col] = df[col].apply(float) / 10000
                        
            ob_position.loc[ob_folder, :] = df[cols].mode().values[0]
            print(df[cols].mode().values[0])
```
